```sql
-- Create the table
CREATE OR REPLACE TABLE workspace.neb.team_stats (
    member_id INT,
    member_name STRING,
    role STRING,
    score INT
);

-- Insert sample data
INSERT INTO workspace.neb.team_stats VALUES 
    (1, 'Simon', 'Lead', 95),
    (2, 'Emma', 'Analyst', 88),
    (3, 'Victor', 'Dev', 92);

-- Run the query
SELECT 
    role, 
    AVG(score) as average_team_score 
FROM workspace.neb.team_stats 
GROUP BY role;
```
